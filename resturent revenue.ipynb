```python
train=pd.read_csv('train.csv')
train = train.drop('Type', axis=1)
train = train.drop('City Group', axis=1)
train = train.drop('City', axis=1)
train = train.drop('Open Date', axis=1)
train = train.drop('Id', axis=1)
```

```python
test=pd.read_csv('test.csv')
test=test.drop(['Type'], axis=1)
test=test.drop(['City'], axis=1)
test=test.drop(['City Group'], axis=1)
test=test.drop(['Open Date'], axis=1)

```

```python
X_train = train.drop("revenue", axis=1)
Y_train = train["revenue"]
X_test  = test.drop("Id", axis=1).copy()
Y_test = test["Id"]
#Y_test = test["Id"]
X_train.shape, Y_train.shape, X_test.shape,Y_test.shape
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC, LinearSVC
from sklearn.ensemble import RandomForestClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.linear_model import perceptron
from sklearn.linear_model import SGDClassifier
from sklearn.tree import DecisionTreeClassifier
```

```python
svc = SVC()
svc.fit(X_train, Y_train)
Y_pred = svc.predict(X_test)
prediction_svc = round(svc.score(X_train, Y_train) * 100, 2)
prediction_svc
```

```python
# Decision Tree

decision_tree = DecisionTreeClassifier()
decision_tree.fit(X_train, Y_train,)
Y_pred = decision_tree.predict(X_test)
prediction_decision_tree = round(decision_tree.score(X_train, Y_train) * 100, 2)
prediction_decision_tree
```

```python
# Random Forest

random_forest = RandomForestClassifier(n_estimators=100)
random_forest.fit(X_train, Y_train)
Y_pred = random_forest.predict(X_test)
random_forest.score(X_train, Y_train)
prediction_random_forest = round(random_forest.score(X_train, Y_train) * 100, 2)
prediction_random_forest
```

```python
submission = pd.DataFrame({
        "Id": test["Id"],
        "Prediction": Y_pred
    })
```
